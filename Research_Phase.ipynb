```python
# prompt: import file from my google drive

import os
breaches = open('/content/drive/MyDrive/Colab Notebooks/DataBreaches(2004-2021).csv.xls',"r")
breaches = breaches.read()
breaches = breaches.split("\n")

count = 0
print(f"data_structure: {breaches[0]}")
for i in range(1,len(breaches)):
  count += 1
print(count)

import pandas as pd
import matplotlib.pyplot as plt


def load_data_breaches(file_path):
    try:
        # Use pandas to read the CSV file
        df = pd.read_csv(file_path)
        return df
    except Exception as e:
        print(f"Error reading file: {e}")
        return None


def aggregate_records_by_year(df):

    yearly_records = df.groupby('Year')['Records'].sum().reset_index()
    return yearly_records

# Create a bar graph of records lost by year
def plot_records_by_year(yearly_records):
    plt.figure(figsize=(12, 6))
    plt.bar(yearly_records['Year'].astype(str), yearly_records['Records'] / 1_000_000)
    plt.title('Records Lost by Year', fontsize=15)
    plt.xlabel('Year', fontsize=12)
    plt.ylabel('Records Lost (Millions)', fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


def main():

    file_path = '/content/drive/MyDrive/Colab Notebooks/DataBreaches(2004-2021).csv.xls'

    # Load the data
    df = load_data_breaches(file_path)

    if df is not None:

        yearly_records = aggregate_records_by_year(df)

        # Print yearly totals
        print("Yearly Records Lost:")
        print(yearly_records)

        # Plot the graph
        plot_records_by_year(yearly_records)

if __name__ == "__main__":
    main()
```
